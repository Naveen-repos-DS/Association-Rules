# Encoding (Transaction Encoder)

```python
from mlxtend.preprocessing import TransactionEncoder

transaction = df.values.tolist();transaction
data = [[ item for item in i if pd.notna(item)]for i in transaction];data


tr = TransactionEncoder()
encoded_data = tr.fit_transform(data)
encoded_data
```

# Apriori Alogorithm

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
encoded_df = pd.DataFrame(encoded_data,columns=tr.columns_);encoded_df
```

```python
freq_item = apriori(encoded_df,min_support=0.03,use_colnames=True);freq_item
```

Here i've used minimum support as 0.03(3%) to get optimal result(Not too high and not too low)

```python
rules = association_rules(freq_item,metric='lift',min_threshold=1);rules
```

```python
rules.sort_values('lift',ascending=False)
```

Interview Questions

1. What is lift and why is it important?

Lift measures how much more likely two items occur together than if they were independent.
It is important because it shows the strength and usefulness of an association rule beyond chance.

Lift > 1 - positive association

Lift = 1 - no association

2. What are support and confidence? How are they calculated?

Support: How frequently an itemset appears in the dataset.


Support(A) = Transactions containing A ​/ Total transactions


Confidence: How often item B appears when item A appears.


Confidence(AnB)= Transactions containing A and B​ / Transactions containing A

3. Limitations or challenges of Association Rule Mining

* Can generate too many rules, many of which are irrelevant.

* Computationally expensive for large datasets.

* Does not imply causation, only correlation.

* Choosing appropriate thresholds for support and confidence can be difficult.
